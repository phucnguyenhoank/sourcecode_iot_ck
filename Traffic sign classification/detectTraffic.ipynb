```python
import tkinter as tk
from tkinter import filedialog, Label, Button
import numpy as np
import cv2
from keras.models import load_model
from PIL import Image, ImageTk

# Load model đã train
model = load_model('my_traffic_classifier.keras')

# Dictionary gán class index sang tên biển báo
classes = {
    1:'Speed limit (20km/h)', 2:'Speed limit (30km/h)', 3:'Speed limit (50km/h)', 
    4:'Speed limit (60km/h)', 5:'Speed limit (70km/h)', 6:'Speed limit (80km/h)', 
    7:'End of speed limit (80km/h)', 8:'Speed limit (100km/h)', 9:'Speed limit (120km/h)', 
    10:'No passing', 11:'No passing veh over 3.5 tons', 12:'Right-of-way at intersection', 
    13:'Priority road', 14:'Yield', 15:'Stop', 16:'No vehicles', 
    17:'Veh > 3.5 tons prohibited', 18:'No entry', 19:'General caution', 
    20:'Dangerous curve left', 21:'Dangerous curve right', 22:'Double curve', 
    23:'Bumpy road', 24:'Slippery road', 25:'Road narrows on the right', 
    26:'Road work', 27:'Traffic signals', 28:'Pedestrians', 29:'Children crossing', 
    30:'Bicycles crossing', 31:'Beware of ice/snow', 32:'Wild animals crossing', 
    33:'End speed + passing limits', 34:'Turn right ahead', 35:'Turn left ahead', 
    36:'Ahead only', 37:'Go straight or right', 38:'Go straight or left', 39:'Keep right', 
    40:'Keep left', 41:'Roundabout mandatory', 42:'End of no passing', 
    43:'End no passing veh > 3.5 tons'
}

# Khởi tạo GUI
top = tk.Tk()
top.geometry('800x600')
top.title('Traffic Sign Classification')
top.configure(background='#CDCDCD')

label = Label(top, background='#CDCDCD', font=('arial', 15, 'bold'))
sign_image = Label(top)

# Khởi tạo camera
cap = cv2.VideoCapture(0)

def classify_from_camera(frame):
    # Chuyển đổi ảnh từ BGR sang RGB
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Tiến hành xử lý đơn giản để kiểm tra nếu có đủ sự thay đổi màu sắc (biển báo)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Nếu không phát hiện được bất kỳ đối tượng nào trong ảnh, bỏ qua
    if len(contours) == 0:
        return
    
    # Chuyển đổi ảnh thành dạng phù hợp với mô hình
    image = Image.fromarray(image)
    image = image.resize((30, 30))  # Resize về kích thước đã huấn luyện
    image = np.array(image)
    image = np.expand_dims(image, axis=0)  # Thêm chiều batch
    
    # Dự đoán biển báo
    pred = model.predict([image])
    pred_class = np.argmax(pred, axis=1)[0]
    sign = classes[pred_class + 1]
    
    # In kết quả phân loại lên console
    print("Detected sign:", sign)
    
    # Cập nhật kết quả lên GUI
    label.configure(foreground='#011638', text=sign)

def update_frame():
    ret, frame = cap.read()  # Đọc ảnh từ camera
    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Đổi sang RGB để Tkinter có thể hiển thị
        img = Image.fromarray(frame)
        img = img.resize((400, 300))  # Resize để hiển thị vừa với cửa sổ Tkinter
        imgtk = ImageTk.PhotoImage(image=img)
        
        # Cập nhật frame trong Tkinter
        sign_image.imgtk = imgtk
        sign_image.configure(image=imgtk)
        
        # Gọi hàm phân loại khi có biển báo rõ ràng trong khung hình
        classify_from_camera(frame)
    
    # Tiếp tục cập nhật frame
    sign_image.after(10, update_frame)

def start_camera():
    update_frame()  # Bắt đầu cập nhật frames từ camera

# Nút bắt đầu camera
start_button = Button(top, text="Start Camera", command=start_camera, padx=10, pady=5)
start_button.configure(background='#364156', foreground='white', font=('arial', 10, 'bold'))
start_button.pack(side=tk.BOTTOM, pady=10)

# Hiển thị hình ảnh từ camera
sign_image.pack(side=tk.BOTTOM, expand=True)
label.pack(side=tk.BOTTOM, expand=True)

# Tiêu đề của ứng dụng
heading = Label(top, text="Know Your Traffic Sign", pady=20, font=('arial', 20, 'bold'))
heading.configure(background='#CDCDCD', foreground='#364156')
heading.pack()

top.mainloop()
# Giải phóng camera khi đóng ứng dụng
cap.release()
```
